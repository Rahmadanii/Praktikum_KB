### Import Package

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read Dataset

```python
df = pd.read_csv("tmdb_movies.csv")
df
```

## 1. Grafik Batang

```python
x = df.original_language
y = df.runtime

plt.bar(x, y,)
plt.title("Dataset Movies")
plt.xticks(rotation=90)
plt.xlabel("Tipe Bahasa")
plt.ylabel("Runtime")
plt.show()
```

```python
df.dtypes
```

## 2. Scatter Plot


```python
sns.scatterplot(data = df, x='original_language', y='runtime')
plt.title("Bahasa ori dan runtime")
plt.xticks(rotation=90)
plt.show()
```

## 3. Grafik Pie

```python
labels = df['original_language'].value_counts().rename_axis('original_language').reset_index(name='hasil')
labels
```

```python
plt.figure(figsize=(8,8), dpi=300)
plt.title("Judul")
plt.pie(df['original_language'].value_counts(), labels=labels['original_language'], autopct='%.1f%%', startangle=0)
plt.show()
```
