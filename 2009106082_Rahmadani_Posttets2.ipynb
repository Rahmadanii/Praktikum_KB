```python
#Jumlah Attribute yg memiliki nilai Null
```

```python
Null = len(df.columns[df.isnull().sum() > 0])
print(Null)
```

```python
#Dictionary
```

```python
totalRecord = len(df)
namaAtribut = df.columns
jumlahObjek = len(df.select_dtypes(include = ['object']).columns)
jumlahAngka = len(df.select_dtypes(include = ['int64', 'float64']).columns)
jumlahNull = len(df.columns[df.isnull().sum() > 0])

dictionary_no1 = {
    "total Record": totalRecord,
    "nama Atribut yang ada": namaAtribut,
    "jumlah Objek": jumlahObjek,
    "jumlah angka": jumlahAngka,
    "jumlah Null": jumlahNull,
}

pd.Series(dictionary_no1)
```
